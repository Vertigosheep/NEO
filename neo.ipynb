```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cmap="viridis", cbar=True)
plt.show()
```

```python
sns.countplot(x="Relative Velocity", data=df)
plt.show()
```

```python
for col in df.select_dtypes(include='object'):
    print(col)
    print(df[col].value_counts())
    print("------")
```

```python
sns.countplot(x="Is Potentially Hazardous", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x="Is Potentially Hazardous", data=df)
plt.title("Hazardous vs Non-Hazardous")
plt.show()
```

```python
sns.countplot(x="Orbiting Body", data=df)
plt.title("Orbiting Body Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=True, linewidths=0.5)
plt.show()
```

```python
sns.boxplot(x="Is Potentially Hazardous", y="Max Diameter (km)", data=df)
plt.show()
```

```python
sns.histplot(df["Relative Velocity (km/s)"], kde=True)
plt.show()
```

```python
sns.boxplot(x="Is Potentially Hazardous", y="Miss Distance (km)", data=df)
plt.show()
```
